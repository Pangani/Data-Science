```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 200)
y1 = np.sin(x)
y2 = np.sqrt(x)

plt.plot(x, y1, '-r')
plt.show()
```

```python
x = np.linspace(10, 50, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x/3)

plt.plot(x, y1, label="Sine")
plt.plot(x, y2, label="Cosine")
plt.plot(x, y3, label="Logarithm")

plt.legend(loc='upper left')
plt.show()
```

```python
mu, sigma = 172, 4
x = mu + sigma * np.random.rand(10000)

plt.hist(x, bins=50, density=True, facecolor="blue")
```

```python
bob = (90, 67, 87, 76)
rich = (80, 80, 47, 66)
milca = (40, 95, 76, 89)

skills = ('Python', 'Java', 'Networking', 'Machine Learning')
```

```python
width = 0.2
index = np.arange(4)
plt.bar(index, bob, width=width,
        label='Bob')
plt.bar(index+width, rich, width=width,
        label='Rich')
plt.bar(index+2*width, milca, width=width,
        label='Milca')

plt.xticks(index+width, skills)
plt.ylim(0, 120)
plt.title("IT skill levels")
plt.ylabel("skill level")
plt.xlabel("IT skills")
plt.legend()
```

```python
labels = ('American', 'German', 'French', 'Other')
values = (47, 23, 20, 10)

plt.pie(values, labels=labels, autopct="%.2f%%",shadow=True)
plt.title("Country of Origin")
plt.show()
```
